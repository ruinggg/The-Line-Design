```python
import win32com.client as win32
from time import time

def CaptureBeamDesignToHeatMap():
    # 設定工作簿路徑 (請確認路徑正確)
    workbook_path = r"C:\Users\jia-hau.liu\Python\20250310_M46_JV3_12Cores_LB_Gravity.xlsm"
    excel = win32.gencache.EnsureDispatch('Excel.Application')
    excel.Visible = True  # 如需觀察 Excel 操作，可設 True
    wb = excel.Workbooks.Open(workbook_path)
    
    # -------------------------------------------------------------
    # Speed Settings：關閉螢幕更新、事件，並設為手動計算
    # -------------------------------------------------------------
    excel.ScreenUpdating = False
    excel.EnableEvents = False
    # xlCalculationManual = -4135, xlCalculationAutomatic = -4105
    excel.Calculation = -4135

    startTime = time()  # 計時開始

    # -------------------------------------------------------------
    # 工作表參考
    # -------------------------------------------------------------
    BeamDesignSht = wb.Sheets("Beam Design")
    HeatMapSht = wb.Sheets("HeatMap")
    StorySht = wb.Sheets("Story Data")
    
    # -------------------------------------------------------------
    # 依故事群進行處理 (k 從 1 到 18)
    # -------------------------------------------------------------
    for k in range(1, 19):
        # 在 HeatMap 的單一儲存格記錄目前故事群 (可依需要修改)
        HeatMapSht.Range(HeatMapSht.Cells(1, 2), HeatMapSht.Cells(1, 2)).Value = k
        
        # ---------------------------------------------------------
        # 設定 Story_Start, Story_End, BDPlace, jEnd
        # ---------------------------------------------------------
        if k == 1:
            Story_Start = StorySht.Cells(2, 10).Value
            Story_End   = StorySht.Cells(2, 11).Value
            BDPlace = 5
            jEnd = BDPlace + 263
        elif k == 2:
            Story_Start = StorySht.Cells(3, 10).Value
            Story_End   = StorySht.Cells(3, 11).Value
            BDPlace = 270
            jEnd = BDPlace + 263
        elif k == 3:
            Story_Start = StorySht.Cells(4, 10).Value
            Story_End   = StorySht.Cells(4, 11).Value
            BDPlace = 535
            jEnd = BDPlace + 263
        elif k == 4:
            Story_Start = StorySht.Cells(5, 10).Value
            Story_End   = StorySht.Cells(5, 11).Value
            BDPlace = 800
            jEnd = BDPlace + 263
        elif k == 5:
            Story_Start = StorySht.Cells(6, 10).Value
            Story_End   = StorySht.Cells(6, 11).Value
            BDPlace = 1065
            jEnd = BDPlace + 263
        elif k == 6:
            Story_Start = StorySht.Cells(7, 10).Value
            Story_End   = StorySht.Cells(7, 11).Value
            BDPlace = 1330
            jEnd = BDPlace + 263
        elif k == 7:
            Story_Start = StorySht.Cells(8, 10).Value
            Story_End   = StorySht.Cells(8, 11).Value
            BDPlace = 1595
            jEnd = BDPlace + 263
        elif k == 8:
            Story_Start = StorySht.Cells(9, 10).Value
            Story_End   = StorySht.Cells(9, 11).Value
            BDPlace = 1860
            jEnd = BDPlace + 263
        elif k == 9:
            Story_Start = StorySht.Cells(10, 10).Value
            Story_End   = StorySht.Cells(10, 11).Value
            BDPlace = 2125
            jEnd = BDPlace + 263
        elif k == 10:
            Story_Start = StorySht.Cells(11, 10).Value
            Story_End   = StorySht.Cells(11, 11).Value
            BDPlace = 2390
            jEnd = BDPlace + 263
        elif k == 11:
            Story_Start = StorySht.Cells(12, 10).Value
            Story_End   = StorySht.Cells(12, 11).Value
            BDPlace = 2655
            jEnd = BDPlace + 263
        elif k == 12:
            Story_Start = StorySht.Cells(13, 10).Value
            Story_End   = StorySht.Cells(13, 11).Value
            BDPlace = 2920
            jEnd = BDPlace + 263
        elif k == 13:
            Story_Start = StorySht.Cells(14, 10).Value
            Story_End   = StorySht.Cells(14, 11).Value
            BDPlace = 3185
            jEnd = BDPlace + 263
        elif k == 14:
            Story_Start = StorySht.Cells(15, 10).Value
            Story_End   = StorySht.Cells(15, 11).Value
            BDPlace = 3450
            jEnd = BDPlace + 263
        elif k == 15:
            Story_Start = StorySht.Cells(16, 10).Value
            Story_End   = StorySht.Cells(16, 11).Value
            BDPlace = 3715
            jEnd = BDPlace + 263
        elif k == 16:
            Story_Start = StorySht.Cells(17, 10).Value
            Story_End   = StorySht.Cells(17, 11).Value
            BDPlace = 3980
            jEnd = BDPlace + 199
        elif k == 17:
            Story_Start = StorySht.Cells(18, 10).Value
            Story_End   = StorySht.Cells(18, 11).Value
            BDPlace = 4181
            jEnd = BDPlace + 263
        elif k == 18:
            Story_Start = StorySht.Cells(19, 10).Value
            Story_End   = StorySht.Cells(19, 11).Value
            BDPlace = 4446
            jEnd = BDPlace + 263
        else:
            wb.Close(False)
            excel.Quit()
            raise Exception("Invalid group value (k={})".format(k))
        
        # ---------------------------------------------------------
        # 逐一處理每個故事 (storyRow 為行號，從 Story_Start 到 Story_End)
        # ---------------------------------------------------------
        for storyRow in range(int(Story_Start), int(Story_End) + 1):
            # 取得故事標籤，並寫入 Beam Design 指定儲存格
            storyLabel = StorySht.Cells(storyRow, 1).Value
            BeamDesignSht.Range(BeamDesignSht.Cells(BDPlace, 3),
                                  BeamDesignSht.Cells(BDPlace, 3)).Value = storyLabel
            
            # 強制重新計算一次 (沿用 Excel 計算引擎)
            excel.Calculate()
            
            # 讀取資料範圍：假設從 BDPlace+4 到 jEnd, 欄位從 5 到 50
            firstDataRow = BDPlace + 4
            rng = BeamDesignSht.Range(BeamDesignSht.Cells(firstDataRow, 5),
                                      BeamDesignSht.Cells(jEnd, 50))
            arrData = rng.Value  # 取得2D陣列 (tuple of tuples)
            rowCount = len(arrData)
            
            # 建立用來儲存各組最大 V_DCR 與相對應的 rho 與 Vmax_DCR (組別 1~24)
            groupMaxVDCR = [0] * 25
            groupRho = [None] * 25
            groupVmax = [None] * 25
            
            # 逐列掃描 arrData (每列為 tuple)
            for r in range(rowCount):
                thisGroup = arrData[r][0]  # 對應工作表的第 5 欄 (在陣列中 index 0)
                if thisGroup is None:
                    continue
                if 1 <= thisGroup <= 24:
                    # 根據註解：
                    #   - rho 在第 44 欄，對應 index = 44 - 5 = 39
                    #   - V_DCR 在第 49 欄，對應 index = 49 - 5 = 44
                    #   - Vmax_DCR 在第 50 欄，對應 index = 50 - 5 = 45
                    thisVDCR = arrData[r][44]
                    if thisVDCR is None:
                        continue
                    try:
                        thisVDCR_num = float(thisVDCR)
                    except Exception:
                        continue
                    if thisVDCR_num > groupMaxVDCR[int(thisGroup)]:
                        groupMaxVDCR[int(thisGroup)] = thisVDCR_num
                        groupRho[int(thisGroup)] = arrData[r][39]
                        groupVmax[int(thisGroup)] = arrData[r][45]
            
            # 將結果寫入 HeatMap 工作表
            # 寫入行為 storyRow+2，依組別寫入：
            #   - ρ 寫入 (storyRow+2, group+2)
            #   - V_DCR 寫入 (storyRow+2, group+28)
            #   - Vmax_DCR 寫入 (storyRow+2, group+54)
            for g in range(1, 25):
                val_rho = groupRho[g] if groupRho[g] is not None else ""
                val_vdcr = groupMaxVDCR[g] if groupMaxVDCR[g] is not None else ""
                val_vmax = groupVmax[g] if groupVmax[g] is not None else ""
                HeatMapSht.Range(HeatMapSht.Cells(storyRow + 2, g + 2),
                                 HeatMapSht.Cells(storyRow + 2, g + 2)).Value = val_rho
                HeatMapSht.Range(HeatMapSht.Cells(storyRow + 2, g + 28),
                                 HeatMapSht.Cells(storyRow + 2, g + 28)).Value = val_vdcr
                HeatMapSht.Range(HeatMapSht.Cells(storyRow + 2, g + 54),
                                 HeatMapSht.Cells(storyRow + 2, g + 54)).Value = val_vmax

    # -------------------------------------------------------------
    # 還原 Excel 設定
    # -------------------------------------------------------------
    excel.Calculation = -4105  # 設定回自動計算
    excel.ScreenUpdating = True
    excel.EnableEvents = True
    
    elapsedTime = time() - startTime
    # 使用 Excel 內建的 MessageBox 顯示總耗時
    
    wb.Save()
    wb.Close()
    excel.Quit()

if __name__ == '__main__':
    CaptureBeamDesignToHeatMap()
```
